```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
final_df = pd.read_pickle("./pickles/combined_dataframe.pkl")
```

# Q: is the pollution (SO2) increasing with the ratio of population growth?

```python
final_df.where((final_df['year']>=2013) & (final_df['year']<2018)).groupby("city").plot(x="quarter",y=["SO_2","ratio"])
```

# Q: is the pollution (PM10) increasing with the ratio of population growth?  

# seasonality 

```python
final_df.groupby("city").plot(x="quarter",y=["PM10","ratio"])
```

## graph formating

```python
# Madrid
import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

# get plotting data
x_mad = final_df["date"].where(final_df["city"] == "madrid")
y_mad = final_df["PM10"].where(final_df["city"] == "madrid")
y_mad_2 = final_df["ratio"].where(final_df["city"] == "madrid")
# define axis and parameters
x = (x_mad)
y = (y_mad)

# create plot
fig = plt.figure(figsize=(10, 6), dpi=80, frameon=True)
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x_mad, y_mad, alpha=0.5, c="green", edgecolors='none', s=20)
plt.ylim(0,250)
ax2 = ax1.twinx()
ax2.plot(final_df["date"].where(final_df["city"] == "madrid"),final_df["ratio"].where(final_df["city"] == "madrid"), color="black")
#plt.ylim(0,200)
# set labels and layout
ax1.set_ylabel("PM10 [μg / m3]", fontsize=12)
ax2.set_ylabel("100km / population", fontsize=12)
plt.grid(True)

#plt.legend(loc=2)
```
