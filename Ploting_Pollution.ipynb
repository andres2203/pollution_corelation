# Importing merged pollution dataframe

```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
pollution_final=pd.read_pickle("./pickles/merged_pollution.pkl")
```

### Creating new column for the year, month and quarter

```python
pollution_final=pollution_final.assign(year=pollution_final['date'].dt.year)
pollution_final=pollution_final.assign(month=pollution_final['date'].dt.to_period('M'))
pollution_final=pollution_final.assign(quarter=pollution_final['date'].dt.to_period('Q'))
```

```python
pollution_final
```

```python
pollution_final=pollution_final.astype({"date":"datetime64","CO":"float","NO_2":"float","NOx":"float", "O_3":"float", "PM10":"float","PM25":"float", "SO_2":"float","city":"object","year":"int64"})
```

```python
pollution_final.isna().sum()/len(pollution_final)*100
```

## Plotting pollution evolution

### PM10

```python
pollution_final.groupby("city").plot(x="quarter",y="PM10")
```

### NO_2

```python
pollution_final.groupby("city").plot(x="quarter",y="NO_2")
```

### O_3

```python
pollution_final.groupby("city").plot(x="quarter",y="O_3")
```

### SO_2

```python
pollution_final.groupby("city").plot(x="quarter",y="SO_2")
```

### Higher cncentration of NO2 in Hong Kong

```python
fig, ax = matplotlib.pyplot.subplots()
pollution_final.groupby("city").plot(x="quarter",y="NO_2",ax=ax)
```
